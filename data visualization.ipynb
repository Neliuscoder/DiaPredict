```python
#import all the necessary libraries 
#to load data from our csv file we are going to use pandas
#for our data visualization we will use matplotlib and seaborn
#I will also import AutoViz to give me cleaning suggestions
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from autoviz import data_cleaning_suggestions

%matplotlib inline

df = pd.read_csv(r"C:\Users\HP\Desktop\diabetes.csv")
#df.sample(10)
data_cleaning_suggestions(df)
df
```

```python
#detecting and removing outliers
#using percentiles
max_thresold=df["DiabetesPedigreeFunction"].quantile(0.95)
max_thresold
#I have chosen my maximum threshold to be at 95 percent.
```

```python
#and my minimum to be at 5 percent 
min_thresold=df["DiabetesPedigreeFunction"].quantile(0.05)
min_thresold
```

```python
#to check for DiabetesPedigreeFunctions that are below the minimum thresold
df[df["DiabetesPedigreeFunction"]<min_thresold]
```

```python
#the above rows are below the maximum threshold and the rows below are above minimum threshold
```

```python
#checking DiabetesPedigreeFunction that are above maximum threshold
df[df["DiabetesPedigreeFunction"]>max_thresold]

```

```python
#to remove outliers i am going to generate values that lie between my minimum and maximum threshold 

df[(df["DiabetesPedigreeFunction"]>min_thresold)& (df["DiabetesPedigreeFunction"]<max_thresold)]
```
